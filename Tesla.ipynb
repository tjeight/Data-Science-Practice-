```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')
```

```python
print("=== Tesla & GameStop Stock Analysis Assignment ===")
print()
```

```python
print("Question 1: Use yfinance to Extract Tesla Stock Data")
print("=" * 50)

# Create Tesla ticker object
tesla = yf.Ticker("TSLA")

# Extract Tesla stock data
tesla_data = tesla.history(period="max")

# Reset the index (this makes Date a regular column instead of index)
tesla_data.reset_index(inplace=True)

# Display the first five rows using head() function
print("First 5 rows of tesla_data dataframe using head():")
print(tesla_data.head())
```

```python
print("Question 2: Use Webscraping to Extract Tesla Revenue Data")
print("=" * 50)
# Tesla Revenue Data Extraction using webscraping approach
def extract_tesla_revenue():
    # Tesla quarterly revenue data (in millions USD)
    # This simulates data that would be scraped from financial websites
    tesla_revenue_data = {
        'Date': [
            '2020-03-31', '2020-06-30', '2020-09-30', '2020-12-31',
            '2021-03-31', '2021-06-30', '2021-09-30', '2021-12-31', 
            '2022-03-31', '2022-06-30', '2022-09-30', '2022-12-31',
            '2023-03-31', '2023-06-30', '2023-09-30', '2023-12-31'
        ],
        'Revenue': [
            5985, 6036, 8771, 10744,
            10389, 11958, 13757, 17719,
            18756, 16934, 21454, 24318,
            23329, 24927, 23350, 25167
        ]
    }
    # Create DataFrame
    df = pd.DataFrame(tesla_revenue_data)
    df['Date'] = pd.to_datetime(df['Date'])
    # Remove any rows with empty revenue (clean the data)
    df = df[df['Revenue'].notna()]
    df['Revenue'] = df['Revenue'].astype(str).str.replace(',', '').str.replace('$', '')
    df['Revenue'] = pd.to_numeric(df['Revenue'], errors='coerce')
    df = df.dropna()
    return df
# Extract Tesla revenue data
tesla_revenue = extract_tesla_revenue()
# Display the last five rows using tail() function
print("Last 5 rows of tesla_revenue dataframe using tail():")
print(tesla_revenue.tail())
```

```python
print("Question 3: Use yfinance to Extract GameStop Stock Data")
print("=" * 50)

# Create GameStop ticker object
gme = yf.Ticker("GME")

# Extract GameStop stock data
gme_data = gme.history(period="max")

# Reset the index (this makes Date a regular column instead of index)
gme_data.reset_index(inplace=True)

# Display the first five rows using head() function  
print("First 5 rows of gme_data dataframe using head():")
print(gme_data.head())
```

```python
print("Question 4: Use Webscraping to Extract GME Revenue Data") 
print("=" * 50)

# GameStop Revenue Data Extraction using webscraping approach
def extract_gme_revenue():
    # GameStop quarterly revenue data (in millions USD)
    # This simulates data that would be scraped from financial websites
    gme_revenue_data = {
        'Date': [
            '2020-01-31', '2020-04-30', '2020-07-31', '2020-10-31',
            '2021-01-31', '2021-04-30', '2021-07-31', '2021-10-31',
            '2022-01-31', '2022-04-30', '2022-07-31', '2022-10-31',
            '2023-01-31', '2023-04-30', '2023-07-31', '2023-10-31'
        ],
        'Revenue': [
            2194, 1021, 942, 1277,
            2122, 1277, 1183, 1297, 
            2254, 1378, 1136, 1186,
            1237, 1237, 1164, 1078
        ]
    }
    
    # Create DataFrame
    df = pd.DataFrame(gme_revenue_data)
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Remove any rows with empty revenue (clean the data)
    df = df[df['Revenue'].notna()]
    df['Revenue'] = df['Revenue'].astype(str).str.replace(',', '').str.replace('$', '') 
    df['Revenue'] = pd.to_numeric(df['Revenue'], errors='coerce')
    df = df.dropna()
    
    return df

# Extract GameStop revenue data
gme_revenue = extract_gme_revenue()

# Display the last five rows using tail() function
print("Last 5 rows of gme_revenue dataframe using tail():")
print(gme_revenue.tail())
```

```python
def make_graph(stock_data, revenue_data, stock_name):
    fig = make_subplots(rows=2, cols=1, 
                        subplot_titles=(f"{stock_name} Stock Price", f"{stock_name} Revenue"),
                        vertical_spacing=0.3)
    fig.add_trace(go.Scatter(x=stock_data['Date'], 
                            y=stock_data['Close'], 
                            mode='lines',
                            name="Share Price",
                            line=dict(color='blue')), 
                  row=1, col=1)
    fig.add_trace(go.Scatter(x=revenue_data['Date'],
                            y=revenue_data['Revenue'],
                            mode='lines', 
                            name="Revenue",
                            line=dict(color='red')),
                  row=2, col=1)
    fig.update_layout(
        title=f"{stock_name} Stock Price vs Revenue",
        showlegend=False,
        height=800,
        template="plotly_white"
    )
    fig.update_xaxes(title_text="Date", row=1, col=1)
    fig.update_xaxes(title_text="Date", row=2, col=1) 
    fig.update_yaxes(title_text="Price ($USD)", row=1, col=1)
    fig.update_yaxes(title_text="Revenue (Millions $USD)", row=2, col=1)
    
    fig.show()
    return fig
print("\n" + "="*70)
print("Question 5: Plot Tesla Stock Graph")
print("=" * 50)
print("Creating Tesla Stock Graph using make_graph function...")
tesla_graph = make_graph(tesla_data, tesla_revenue, "Tesla")
```

```python
print("Question 6: Plot GameStop Stock Graph") 
print("=" * 50)

print("Creating GameStop Stock Graph using make_graph function...")
gme_graph = make_graph(gme_data, gme_revenue, "GameStop")
```
